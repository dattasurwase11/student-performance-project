```python
# Calculate and show performance metrics
# Use common regression metrics like:
# Mean Absolute Error (MAE)
# Mean Squared Error (MSE) or Root MSE (RMSE)
# R-squared (R²)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(test_data['math score'], predictions)
rmse = np.sqrt(mean_squared_error(test_data['math score'], predictions))
r2 = r2_score(test_data['math score'], predictions)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")
```

```python
# Scatter plot of Actual vs Predicted scores
# Residual plot (shows prediction errors)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Actual and predicted
actual = test_data['math score']
predicted = predictions

# Scatter plot: Actual vs Predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=actual, y=predicted, alpha=0.7)
plt.plot([actual.min(), actual.max()], [actual.min(), actual.max()], 'r--')  # Diagonal line y=x
plt.xlabel('Actual Math Score')
plt.ylabel('Predicted Math Score')
plt.title('Actual vs Predicted Math Scores')
plt.grid(True)
plt.show()

# Residuals calculation
residuals = actual - predicted

# Residual plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=predicted, y=residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')  # Zero error line
plt.xlabel('Predicted Math Score')
plt.ylabel('Residuals (Actual - Predicted)')
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```

```python
# Explanation:
The scatter plot shows how close predicted values are to actual values.
The red dashed line is the perfect prediction line (y = x). The closer points cluster to it, the better.
The residual plot shows errors — residuals ideally scatter randomly around zero with no clear pattern. This suggests no bias.
```
